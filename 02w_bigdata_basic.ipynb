```python
# 1. 0부터 9까지 숫자 중 짝수의 제곱만을 포함하는 리스트 생성
# 기본 방식
squares_basic = []
for x in range(10):
    if x % 2 == 0:
        squares_basic.append(x**2)

# 리스트 컴프리헨션 방식 (코드 간결화 및 성능 향상)
squares_comp = [x**2 for x in range(10) if x % 2 == 0]

print(f"기본 방식 결과: {squares_basic}")
print(f"리스트 컴프리헨션 결과: {squares_comp}")

# 2. 리스트 내 문자열의 길이를 구하는 예제
words = ['Python', 'BigData', 'AI', 'Structure']
lengths = [len(word) for word in words]
print(f"단어 길이 리스트: {lengths}") # [6, 7, 2, 9]
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}
print(A.union(B)) # 합집합
print(A.intersection(B)) # 교집합
print(A.difference(B)) # 차집합
print(B.difference(A)) # 차집합
print(A.symmetric_difference(B)) # 여집합
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}
print(A | B) # 합집합
print(A & B) # 교집합
print(A - B) # 차집합
print(B - A) # 차집합
print(A ^ B) # 여집합
```
